```python
import numpy as np 
import numpy.linalg as npla 

import scipy 
from scipy import sparse
```

```python
A = np.array([[4,-1,-1],[-2,6,1],[-1,1,7]])
b = np.array([3,9,-6])
xideal  = np.array([1,2,-1])
```

```python
m,n = A.shape 
d = A.diagonal()

#.diag is used to turn the diagonals vectorinto the diagonal matrix D 
#way to turn vector of a n x 1 size to n x n matrix
D = np.diag(d)

C = A - D
print("c = \n", C, "\n")

#function to make a vector with size n x 1by n matrix with all zeros 
x = np.zeros(n)
```

```python
xnew = (b -C@x)/d
print("x[1] = ", xnew)

#following lines only if x solution is known 
relres = npla.norm(xnew - xideal) / npla.norm(xideal)
error = npla.norm(xnew-xideal)
print("error: ", error, "\nrelres:", relres)

#otherwise would use 
error = npla.norm(A @ xnew - b) #ideally would be 0 if xnew is = x 
print("error: ", error)
relres = npla.norm(A @ xnew - b)/npla.norm(b)
print("relres: ", relres)
```

```python
x = [0,0,0]
for i in range(70):
    #print("b-C@x: ", b - C@x)
    #print("d: ", d)
    xnew = (b -C@x)/d
    print("x[", i, "] = ", xnew, end = " ")

    #relres = npla.norm(xnew - xideal) / npla.norm(xideal)
    error = npla.norm(xnew-xideal)
    print("error: ", error)
    #print("\nrelres:", relres)
    x = xnew
    if error <= 1e-8:
        break 
```
