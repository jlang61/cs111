```python
import numpy as np 
import numpy.linalg as npla 

import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import axes3d
date = np.array(range(1900,2030,10)) - 1900
population = 1000 * np.array([ 75995, 91972, 105711, 123203, 131669,
150697, 179323, 203212, 226505, 249633,
281422, 308746, 329168])
t = date/10
print(t)
y = population

m = len(y) 
A = np.ones((m,7))

A[:,1] = t
A[:,2] = t ** 2 
A[:,3] = t ** 3
A[:,4] = t ** 4
A[:,5] = t ** 5
A[:,6] = t ** 6
# print(A)
x = npla.lstsq(A,y, rcond= None)[0]
print(x) 

%matplotlib inline
plt.figure()
plt.plot(t,y,'.')
# print(x)
polyt = np.linspace(0,12,num=100)
polyy = x[0] + x[1]*polyt + x[2]*polyt**2 + x[3] * polyt **3 + x[4] * polyt ** 4 + x[5] * polyt ** 5 + x[6] * (polyt ** 6)

plt.plot(polyt,polyy)
plt.xlabel('decades since 1990')
plt.ylabel('population')
plt.legend(('observations', '6th degree fit'))
d = 12 
print(x[0] + d * x[1] + (d **2) * x[2] + (d ** 3) * x[3] + (d ** 4) * x[4] + (d ** 5) * x[5] + (d ** 6) * x[6])

d = 13
print(x[0] + d * x[1] + (d **2) * x[2] + (d ** 3) * x[3] + (d ** 4) * x[4] + (d ** 5) * x[5] + (d ** 6) * x[6])
print('The relative residual norm:\n', npla.norm(y - A@x)/npla.norm(y))
# print(A@x)
# print(y)
# print(y - A @x)
# print(npla.norm(y))
```
